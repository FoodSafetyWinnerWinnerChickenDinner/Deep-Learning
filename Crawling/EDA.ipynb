# Import Library and Load Data 

```python
import pandas as pd
import numpy as np
from glob import glob
```

```python
header = ['음식 이름', '지역 이름', '브랜드', '총 내용량', '분류', '열량', '탄수화물', '단백질', '지방', '당류', '나트륨', '콜레스테롤', '포화지방산', '트랜스지방', '기타']
```

```python
filelists = glob(f'../Data/output/*.txt')
```

```python
for i in range(len(filelists)):
    try:
        print('='*50)
        print(f'File Name : {filelists[i]}')
        print(pd.read_csv(filelists[i], sep=';', header=None).head(1))
    except:
        print('\n')
        print('*'*50)
        print(f'Parse Error : {filelists[i]}')
        print('*'*50)
        print('\n')
```

```python
df = pd.DataFrame()
cnt = 0
for file in filelists:
    df_ = pd.read_csv(file, sep=';', header=None)
    df_.columns = header
    df_ = df_.iloc[:, :-1]
    df = pd.concat([df, df_])
    cnt += 1
```

```python
df.head()
```

```python
df = df.dropna(subset=['음식 이름'])
df[df['음식 이름'].isna()]
```

```python
print(cnt)
df.head()
```

```python
foods = ['치킨', '돼지', '구이', '회', '라면', '아메리카노', '쥬스', '빵', '피자', '김치', '밥', '탕', '떡볶이', '순대', '튀김']
for food in foods:
    print(f'음식 이름 : {food}')
    print(df[df['음식 이름'].str.contains(food)].head())
    print('='*50)
```

# Null 제거 

```python
df = df.dropna(subset=['열량', '탄수화물', '단백질', '지방', '당류', '나트륨', '콜레스테롤', '포화지방산', '트랜스지방'])
df.head()
```

# 치킨 

```python
food = '치킨'
print(f'음식 이름 : {food}')
chicken = df[df['음식 이름'].str.contains(food)]
chicken = chicken.dropna(subset=['분류'])
chicken = chicken[chicken['분류'].str.contains('육')]
summary = chicken.describe()
print(summary.mean().to_dict())
```
